```python
##Dependant Sample :Same Group(FeMale) but Different Condition(etest_p,mba_p)
from scipy.stats import ttest_rel
df=df.dropna()
Female=df[df['gender']=="F"]['etest_p']
Female1=df[df['gender']=="F"]['mba_p']
print ("FeMale:significance level (Î±)")
ttest_rel (Female,Female1)
```

```python
#STANDARD DISTRIBUTION
import seaborn as sns
df_filtered = df[df['salary'] != 0]
mean=df['salary'].mean()
std=df['salary'].std()
z_score=(df_filtered['salary']-mean)/std
sns.distplot(z_score,kde=True)
plt.title('Standardized Salary Distribution')
plt.xlabel('Standardized Salary')
plt.ylabel('Frequency')
plt.show()
print(f"MEAN: {mean} Standard Deviation: {std}")
```

```python
#Question No.10:What is the probability Density Function of the salary range from 700000 to 900000?
```

```python
def get_pdf_probability(df,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax = sns.distplot(df,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # generate a sample
    sample = df
    # calculate parameters
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # define the distribution
    dist = norm(sample_mean, sample_std)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(df["salary"],700000,900000)
```
