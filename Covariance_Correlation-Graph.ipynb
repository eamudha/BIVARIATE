```python
dataset.drop('sl_no',inplace=True,axis=1)
dataset
```

```python
def QuanQual(dataset):
    Quan=[]
    Qual=[]
    for ColumnName in dataset.columns:
        #print(columnName)
        if(dataset[ColumnName].dtype=='O'):
            #print("qual")
            Qual.append(ColumnName)
        else:
            #print("quan")
            Quan.append(ColumnName)
    return Quan,Qual
```

```python
QuanQual(dataset)
```

```python
Quan,Qual=QuanQual(dataset)
```

```python
print(Quan)
print(Qual)
```

```python
Quandataset=dataset[Quan]
Quandataset
```

```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datasetQuan = pd.DataFrame(Quandataset,columns = Quan)
```

```python
x= datasetQuan['hsc_p']  # Selecting 'hsc_p' column
y = datasetQuan['etest_p']  # Selecting 'etest_p' column
correlation = y.corr(x)  # Calculating correlation between x and y
plt.title('Correlation between hsc_p and etest_p')
plt.scatter(x, y)  # Scatter plot
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')  # Adding regression line
plt.xlabel('hsc_p')
plt.ylabel('etest_p')
plt.show()
```
